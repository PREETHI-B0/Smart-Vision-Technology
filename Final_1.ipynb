```python
import numpy as np
import pandas as pd
import ast
import plotly.express as px
from plotly import graph_objects as go
```

```python
df = pd.read_csv("/content/flipkart_com-ecommerce_sample.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df["retail_price"].fillna(df["retail_price"].median(),inplace=True)
df["discounted_price"].fillna(df["discounted_price"].median(),inplace=True)
```

```python
x=df['retail_price']-df['discounted_price']
y=(x/df['retail_price'])*100
df['discount_percentage']=y
```

```python
df['timestamp']=pd.to_datetime(df['crawl_timestamp'])
df['Time']=df['timestamp'].apply(lambda x : x.time)
df['date']=df['timestamp'].apply(lambda x : x.date)
df.drop(['crawl_timestamp'], axis = 1,inplace=True)
df['main_category']=df['product_category_tree'].apply(lambda x :x.split('>>')[0][2:len(x.split('>>')[0])-1])
```

```python
n = 10
top_products=pd.DataFrame(df['main_category'].value_counts()  [:n]).reset_index()
top_products.rename(columns = {'index':'Top_Products','main_category':'Total_Count'}, inplace = True)

#Top 10 main brands being purchased

n = 10
top_brands=pd.DataFrame(df['brand'].value_counts()[:n]).reset_index()
top_brands.rename(columns = {'index':'Top_Brands','brand':'Total_Count'}, inplace = True)
```

```python
df_discount=df.query('discount_percentage > 90')  #targeting brands giving high discounts
df_discount=df_discount.dropna() #dropping rows with NA values
df_discount["brand"].replace('FashBlush','Fash Blush',inplace=True) #handling spelling errors
max_discount=pd.DataFrame(df_discount.groupby('brand')[['discount_percentage']].mean().sort_values(by=['discount_percentage'],ascending=False).reset_index())
```

```python
px.bar(max_discount, x= 'brand', y='discount_percentage',color='brand',color_discrete_sequence=px.colors.qualitative.Dark2)
```

```python
df_customer=df.groupby("uniq_id")[["discounted_price"]].sum().sort_values(by=['discounted_price'],ascending=[False]).reset_index()

#Top 20 customers spending the most
list1=df_customer[:20]

#plotting a bar graph
px.bar(list1, x= 'uniq_id', y="discounted_price",color='discounted_price',color_continuous_scale=px.colors.diverging.BrBG)
```

```python
# 5 star rating

total_prod=len(df['pid'])  #total products using pid variable
total_ratings=len(df[df['product_rating']!='No rating available']) #total rated products
top_ratings=len(df[df['product_rating']=='5']) #5 star rated products
df_funnel_1 = dict(
    number=[total_prod,total_ratings,top_ratings],
    stage=["Total Products","Products with ratings","Products with 5 star rating"])
funnel_1_fig = px.funnel(df_funnel_1, x='number', y='stage')
funnel_1_fig.show()
```

```python
# import the libraries:

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# take a copy:

cdf = df.copy()
```

```python
# check the nulls and column types:

print(cdf.info())
null_data = cdf.isna()
null_data.sum()
```

```python
plt.figure(figsize =(10,8))
sns.heatmap(null_data)
```

```python
# drop the null rows were 'retail_price' & 'discounted_price':

cdf = cdf.dropna(subset='retail_price')
cdf = cdf.dropna(subset='discounted_price')
cdf.isna().sum()
```

```python
cdf.shape
```

```python
# What are the average top 10 retail and discounted prices of products?

main_grouped_retail = cdf.groupby('main_category')['retail_price'].mean()
main_grouped_discounted = cdf.groupby('main_category')['discounted_price'].mean()
sorted_retail_grouped = main_grouped_retail.sort_values(ascending=False)
sorted_discounted_grouped = main_grouped_discounted.sort_values(ascending=False)

pd.DataFrame(sorted_retail_grouped.head(20))
```

```python
pd.DataFrame(sorted_discounted_grouped)
```

```python
# How do prices vary across different product categories?

group_products_mean = cdf.groupby('main_category')['retail_price'].mean()
group_products_median = cdf.groupby('main_category')['retail_price'].median()

group_products_mean.plot(kind='bar',figsize=(42,12),color='red')
group_products_median.plot(kind='bar',figsize=(42,12),color='blue')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze price distribution by category
price_stats = cdf.groupby('main_category')['retail_price'].describe()
pd.DataFrame(price_stats)
```

```python
# Are there any significant price differences between FK Advantage products and others?
# retail prices:

grouped_FK_retail = cdf.groupby('is_FK_Advantage_product')['retail_price'].describe()
grouped_FK_retail
```

```python
# discounted prices:

grouped_FK_discounted = cdf.groupby('is_FK_Advantage_product')['discounted_price'].describe()
grouped_FK_discounted
```

```python
## Product popularity:
# Which product categories have the most products?

group_count_main_cat = cdf.groupby('main_category')['main_category'].value_counts()
sorted_group_count_main_cat = group_count_main_cat.sort_values(ascending=False)
pd.DataFrame(sorted_group_count_main_cat.head(20))
```

```python
# What are the top 10 most frequently occurring brands?

brand_grouping_count = cdf.groupby('brand')['brand'].value_counts()
sorted_brand_grouping_count = brand_grouping_count.sort_values(ascending=False)
pd.DataFrame(sorted_brand_grouping_count.head(10))
```

```python
# Which products have the highest overall ratings?
cdf['overall_rating'].value_counts()
```

```python
# change the type into float and remove 'no available rating string' and convert it into 0

cdf['overall_rating'] = cdf['overall_rating'].replace('No rating available',0)
cdf['overall_rating'] = cdf['overall_rating'].astype('float64')
cdf.info()
```

```python
# Are there any relationships between product ratings and prices?
cdf['product_rating'] = cdf['product_rating'].replace('No rating available',0)
cdf['product_rating'] = cdf['product_rating'].astype('float')

cdf_rating_sorted = cdf.sort_values('product_rating',ascending=False)
corr_matrix = cdf_rating_sorted[['product_rating','retail_price']].corr()

sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',)
plt.figure(figsize=(42,20))
plt.show()
```
